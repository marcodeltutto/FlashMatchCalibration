```python
fname_tree     = "./flashmatchcalib_output_merged.root"
fname_spectree = "./flashmatchcalib_output_merged.root"
raw_df1 = pd.DataFrame(root2array(fname_tree,'flashmatchcalib/tree'))
raw_df2 = pd.DataFrame(root2array(fname_spectree,'flashmatchcalib/spectree'))

raw_df = pd.concat([raw_df1, raw_df2], axis=1)
df = raw_df.query('not score == inf')

#df = raw_df.query('-20 < ((tpc_xmin - t0*0.11) - qll_xmin) and ((tpc_xmin - t0*0.11) - qll_xmin) < 20')

#df = df.query('run == 5134 and subrun == 33 and event == 1697')
#df = df.query('run == 6336 and subrun == 82 and event == 4114')
#df = df.query('run == 5726 and subrun == 89 and event == 4469')
#df = df.query('run == 5607 and subrun == 63 and event == 3152')
df = df.query('run == 5211 and subrun == 9 and event == 490')

hypo_v = df.hypo_spec.values
flash_v = df.flash_spec.values

print df.score.values
print (df.tpc_xmin.values - df.t0.values*0.11) - df.qll_xmin.values
print (df.tpc_xmin.values)
print df.qll_xmin.values
print df.t0.values
print df.columns
```

```python
event = 0

run_v = df.run.values
subrun_v = df.subrun.values
event_v = df.event.values
score_v = df.score.values
t0_v = df.t0.values
xdiff_v = (df.qll_xmin.values - (df.tpc_xmin.values - df.t0.values*0.11))

print 'run',run_v[event]
print 'subrun',subrun_v[event]
print 'event',event_v[event]
print 'score',score_v[event]
print 't0',t0_v[event]
print 'xdiff',xdiff_v[event]



data_x=[]
data_y=[]  # reco matched flash
data_y2=[] # hypo flash
data_y3=[] # mucs flash

for pmt in xrange(32):
    data_x.append(pmt)
    hypo = hypo_v[event][pmt]
    flash = flash_v[event][pmt]
    data_y.append(hypo)
    data_y2.append(flash)

fig = plt.figure(figsize=(12,8),facecolor='w')
ax = fig.add_subplot(1,1,1)
ax.errorbar(data_x,data_y, fmt='o',label='Hypo flash',markersize=10,linewidth=2, linestyle='-')
ax.legend(numpoints=1, loc='upper left')
ax2 = fig.add_subplot(1,1,1)
ax2.errorbar(data_x,data_y2, fmt='o',label='Reco matched flash',markersize=10,linewidth=2, linestyle='-')
ax2.legend(numpoints=1, loc='upper left')
ax3 = fig.add_subplot(1,1,1)

#plt.ylim(0,800)
plt.tick_params(labelsize=20)
plt.grid()
plt.ylabel('PE Count',fontsize=20,fontweight='bold')
plt.xlabel('PMT ID',fontsize=20,fontweight='bold')
leg=plt.legend(fontsize=20,loc=2)
leg_frame=leg.get_frame()
leg_frame.set_facecolor('white')
plt.title('',fontsize=24,fontweight='bold')
plt.xlim(-1,32)
plt.show()
```
